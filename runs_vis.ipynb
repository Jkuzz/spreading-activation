```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

runs_data = pd.read_csv('runs.csv')
runs_data['Duration'] = runs_data['Duration'].apply(lambda x: float(x[:-3]))
```

```python
configs = ['Actor','CountryOfOrigin','Director','Editor','Genre','Producer','Writer','activation_threshold','decay_factor','fromDecade','inLanguage']

for prop in configs:
    plt.figure()
    sns.regplot(
        data=runs_data, x=prop, y='ndcg',
        scatter_kws={'s': runs_data['Duration']}
    )
```
